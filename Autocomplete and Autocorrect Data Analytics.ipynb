### 2. NLP Preprocessing: Clean and Prepare Data for Analysis

```python
def preprocess_text(text):
    # Tokenize and remove stopwords
    tokens = nltk.word_tokenize(text.lower())
    stop_words = set(nltk.corpus.stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words and word.isalpha()]
    return " ".join(filtered_tokens)

# Preprocess the 'transaction_description' column in the dataset
cleaned_data = [preprocess_text(text) for text in data['transaction_description']]
```

### 3.Autocomplete: Implement an autocomplete algorithm based on n-grams (Bigram model)

```python
class Autocomplete:
    def __init__(self, corpus):
        self.corpus = corpus
        self.model = self.build_ngram_model(corpus)

    def build_ngram_model(self, corpus):
        ngram_model = {}
        for sentence in corpus:
            tokens = sentence.split()
            for i in range(len(tokens) - 1):
                prefix = tokens[i]
                suffix = tokens[i + 1]
                if prefix not in ngram_model:
                    ngram_model[prefix] = []
                ngram_model[prefix].append(suffix)
        return ngram_model

    def predict(self, prefix):
        if prefix in self.model:
            return random.choice(self.model[prefix])  # Randomly pick a next word
        else:
            return None  # No prediction found

autocomplete = Autocomplete(cleaned_data)
print("Autocomplete prediction for 'transaction':", autocomplete.predict("transaction"))
```

### 4. Autocorrect: Implement a spelling correction algorithm using fuzzy matching

```python
def autocorrect(input_text, dictionary):
    suggestions = [(word, fuzz.ratio(input_text, word)) for word in dictionary]
    suggestions.sort(key=lambda x: x[1], reverse=True)
    return suggestions[0][0]  # Return word with the highest similarity score

word_dict = ["transaction", "amount", "payment", "credit", "fraud", "service"]
input_word = "tranction"  # Example of a misspelled word
corrected_word = autocorrect(input_word, word_dict)
print(f"Autocorrected '{input_word}' to '{corrected_word}'")
```
